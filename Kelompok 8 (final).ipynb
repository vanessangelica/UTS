## Normalisasi Data

```python
from sklearn.preprocessing import MinMaxScaler
X = train.drop('Credit_Score', axis=1)  # Kolom fitur
y = train['Credit_Score']  # Kolom target

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)
train_normalized = pd.DataFrame(X_normalized, columns=X.columns)
train_normalized['Credit_Score'] = y
train_normalized=pd.concat([train_normalized.drop(kolom_1, axis=1),train_1], axis=1)
train_normalized
```

```python
#Outlier adalah titik data yang nilainya secara signifikan berbeda dengan populasi tertentu.
#Titik data yang berada di bawah Q1 – 1.5 IQR atau di atas Q3 + 1.5 IQR adalah outlier.

def hapus_pencilan_iqr(train):
    train_clean = train.copy()
    for kolom in train.columns:
        Q1 = train[kolom].quantile(0.25)
        Q3 = train[kolom].quantile(0.75)
        IQR = Q3 - Q1
        batas_bawah = Q1 - 1.5 * IQR
        batas_atas = Q3 + 1.5 * IQR
        train_clean = train_clean[(train_clean[kolom] >= batas_bawah) & (train_clean[kolom] <= batas_atas)]
    return train_clean
train_clean= hapus_pencilan_iqr(train_normalized)
```
